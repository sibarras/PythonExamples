```python
# Para encontrar palindromes en la lista de palabras
def signature(word):
    return ''.join(sorted(word))

def find_anagram(my_word, words):
    my_signature = signature(my_word)

    for word in words:
        if my_signature == signature(word):
            print(word)
```

```python
find_anagram('dictionary', words)
```

```python
%time find_anagram('dictionary', words)
```

```python
# To do this without loops. Use dictionaries
import collections
words_by_sig = collections.defaultdict(set)
for word in words:
    words_by_sig[signature(word)].add(word)
words_by_sig
```

```python
anagrams_by_sig = {sig: wordset for sig, wordset in words_by_sig.items() if len(wordset) > 1}
anagrams_by_sig
```

```python
def find_anagram_fast(my_word:str):
    sig = signature(my_word)
    try:
        return anagrams_by_sig[sig]
    except KeyError:
        return set()
```

```python
%time find_anagram_fast('dictionary')
```

```python
find_anagram_fast('hanna')
```
