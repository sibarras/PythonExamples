```python
word_file = 'files/Exercise Files/chapter3/words.txt'
words = {line.strip().lower() for line in open(word_file, 'rt')}

sig = lambda word: (word if word>word[::-1] else word[::-1])
# With true palindromes
palindromes_list = list({sig(word): (word, word[::-1]) for word in words if word[::-1] in words and len(word)>1}.values())
result = list(sorted(palindromes_list, key=lambda p: p[0]))

# to remove true palindromes
result = {(val1, val2) for val1, val2 in result if val1 != val2}
print(len(result))
result
```
