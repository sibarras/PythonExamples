```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import math, collections
```

```python
import pathlib
from os import path
words_file = pathlib.Path('./files/Exercise Files/chapter3/words.txt')
assert words_file.exists()
```

```python
words = []
for line in open(words_file, 'rt'):
    words.append(line)

len(words)
```

```python
words[:10]
```

```python
# to do it better
words = []
for word in open(words_file, 'rt'):
    words.append(word.strip().lower())
words[:10]
```

```python
# to avoid duplicates
words = set()
for line in open(words_file, 'rt'):
    words.add(line.strip().lower())
```

```python
# Do set comprehension
words = {line.strip().lower() for line in open(words_file, 'rt')}
words
```

```python
# words sorted
words = sorted({line.strip().lower() for line in open(words_file, 'rt')})
words
```
